### Задание 1

Вам надо написать программу, которая спрашивает у пользователя три числа (a, b и c), а дальше сообщает пользователю, обладают или не обладают введённые числа следующими свойствами:

- `c` является решением линейного уравнения `ax + b = 0` (т.е. `ac + b = 0`)
- `a` поделить на `b` равно `c`

Важно! Ваша программа не должна ломаться, если одно из введённых чисел является нулём.

Пример вывода для введённых чисел 42, 7, 6:

```
6 не является решением линейного уравнения 42x + 7 = 0
46 поделить на 7 равно 6
```

````python

def main(a, b, c):
  first_answer = 'не является'
  if a * c + b == 0:
    first_answer = 'является'

  print(str(c) + ' ' + first_answer + ' решением линейного уравнения ' + str(a) + 'x + ' + str(b) + ' = 0')

  second_answer = 'не равно'

  if b != 0 and (a / b) == c:
    second_answer = 'равно'

  print(str(a) + ' поделить на ' + str(b) + ' ' + second_answer + ' ' + str(c))


a = int(input("Введите число a: "))
b = int(input("Введите число b: "))
c = int(input("Введите число c: "))

main(a, b, c)
````

### Задание 2

В одном пироге 4 кусочка. Программа следит за поеданием пирогов: в начале спрашивает, сколько вы планируете съесть кусочков, затем выдаёт вам кусочки, а после каждого целого пирога спрашивает, не наелись ли вы, не стоит ли прекратить вас угощать.

Пример:

````python
num_cakes = 0
num_slices = 0

slices_to_eat = int(input("Сколько хотите кусочков?"))

while num_slices < slices_to_eat:
  print("Вот вам кусочек")
  num_slices += 1

  if num_slices == slices_to_eat:
    print("Вот, вроде накормили вас!")
    break

  if num_slices % 4 == 0:
    num_cakes += 1
    print("Уже съедено пирогов: ", num_cakes)
    answer = input("Ещё? (да / не, спасибо)")
    if answer.lower() == "нет" or answer.lower() == "спасибо":
      print("Вот, вроде накормили вас!")
      break
    else:
      print("Продолжаю угощать!")

print("Приходите еще!")
````

### Задание 3

Саша всю ночь готовится к игре в города, выписывая все встречающиеся в интернете названия городов. Напишите программу, которая принимает на вход строку с названиями городов, затем одну заглавную букву, и выводит все города из строки, которые на нее начинаются.

ВВОД
```
Строка, которая содержит названия городов, разделенные запятой и пробелом.
Заглавная буква.
```
ВЫВОД
```
Города из введенной строки, начинающиеся на заданную заглавную букву, каждое название города на новой строке. Если таких городов нет, то программа не должна ничего выдавать.
```
**Пример:**

Ввод
```
Москва, Киев, Токио, Сеул, Казань, Тверь, Брно, Цюрих, Леон, Малага, Лиссабон, Осло, Ливерпуль
Л
```
Результат
```
Леон
Лиссабон
Ливерпуль
```

````python
cities = input().split(", ")
letter = input()


for city in cities:
  if city[0] == letter:
    print(city)
````
