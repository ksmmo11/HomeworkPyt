### ФИО
Напишите функцию `isfullname`, которая проверяет, является ли строка ФИО.

ФИО - три последовательности буквенных символов через пробел, каждая из которых с большой буквы.

Функция принимает строку и выдаёт либо True, либо False.
```
Халова Камилла Григорьевна -> True
Таня Казакова -> False
Введенский Александр Иван0вич -> False
очень скромный саша -> False
Валдис ? Пельш -> False
```

````python
def isfullname(s):
    words = s.split()
    if len(words) != 3:
        return False
    for word in words:
        if not word.isalpha():
            return False
    return True
````
